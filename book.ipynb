```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('loan_data.csv')
```

```python
df.head()
```

```python
df.rename(columns={
    'person_age': 'Age',
    'person_gender': 'Gender',
    'person_education': 'Education_Level',
    'person_income': 'Annual_Income',
    'person_emp_exp': 'Employment_Experience',
    'person_home_ownership': 'Home_Ownership',
    'loan_amnt': 'Loan_Amount',
    'loan_intent': 'Loan_Purpose',
    'loan_int_rate': 'Interest_Rate',
    'loan_percent_income': 'Income_Percent_for_Loan',
    'cb_person_cred_hist_length': 'Credit_History_Length',
    'credit_score': 'Credit_Score',
    'previous_loan_defaults_on_file': 'Previous_Defaults',
    'loan_status': 'Loan_Status'
}, inplace=True)
```

```python
df.describe().loc[['mean', '50%', 'std']]
```

```python
income_95 = df['Annual_Income'][df['Annual_Income'] <= df['Annual_Income'].quantile(0.95)]
plt.figure(figsize=(10, 6))
plt.hist(income_95, bins=30, edgecolor='black')
plt.title('Distribution of Annual Income (Excluding Top 5% Outliers)')
plt.xlabel('Annual Income')
plt.ylabel('Frequency')
plt.show()
```

```python
education_counts = df['Education_Level'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(education_counts, labels=education_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Education Level')
plt.show()
```
